###Criando os dados a partir das informações obtidas anteriormente

```python
#Agrupando a partir da raça/etnia dos alunos:
df2 = df.groupby('race/ethnicity')[['math score']].mean()
#Apresentando a nova tabela:
df2
```

```python
#Criação do gráfico:
df2.plot(kind = 'line', color = '#6e3c1d')
#Estilizando o gráfico:
plt.title('Média das notas de matemática por raça/etnia dos alunos')
plt.xlabel('Raça/Etnia')
plt.ylabel('Médias das notas de matemática')
```

```python
#Agrupando a partir do Nível de Escolaridade dos pais:
df3 = df.groupby('parental level of education')[['writing score']].mean()
#Apresentando a nova tabela:
df3
```

```python
#Criação do gráfico:
df3.plot(kind = 'barh', color='#5e69ff')
#Estilizando o gráfico:
plt.title('Média das notas de matemática por raça/etnia dos alunos')
plt.xlabel('Nota média em Escrita')
plt.ylabel('Nível de escolaridade dos pais')
```

```python
#Agrupando a partir da participação em um curso preparatório para os testes:
df4 = df.groupby('test preparation course')[['reading score']].mean()
#Apresentando a nova tabela:
df4
```

```python
#Criação do gráfico:
df4.plot(kind = 'pie', subplots=True, colors=['#81f542','#f4ff24'])
#Estilizando o gráfico:
plt.title('Média das notas de matemática por raça/etnia dos alunos')
```
